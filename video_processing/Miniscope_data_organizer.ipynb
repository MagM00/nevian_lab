# Rename files

```python
import os

path = r'H:\Miniscope_data\ACC_formalin_SP_data'

for dirpath, dirnames, filenames in os.walk(path):
    for filename in filenames:
        if filename.endswith(".avi") and not filename.startswith("msCam"):
            new_name = 'msCam' + filename
            os.rename(os.path.join(dirpath, filename), os.path.join(dirpath, new_name))
            print(f"Renamed {filename} to {new_name} in folder {dirpath}")
```

# Move files
- Traverse the directory structure to find folders named YYYY_MM_DD.
- Inside each such folder, look for subfolders with the pattern HH_MM_SS\Miniscope.
- If the respective SP folder doesn't exist for the YYYY_MM_DD folder, create it.
- Move (cut and paste) all data inside the Miniscope folder to the SP folder.

```python
import os
import re
import shutil

path = r'H:\Miniscope_data\ACC_formalin_SP_data'

# Regular expression pattern to identify folders
date_pattern = re.compile(r'\d{4}_\d{2}_\d{2}')
time_pattern = re.compile(r'\d{2}_\d{2}_\d{2}')

for dirpath, dirnames, filenames in os.walk(path):
    for dirname in dirnames:
        if date_pattern.match(dirname):  # Found a YYYY_MM_DD folder
            for sub_dirpath, sub_dirnames, sub_filenames in os.walk(os.path.join(dirpath, dirname)):
                for sub_dirname in sub_dirnames:
                    if time_pattern.match(sub_dirname):  # Found a HH_MM_SS folder inside YYYY_MM_DD folder
                        miniscope_path = os.path.join(sub_dirpath, sub_dirname, 'Miniscope')
                        if os.path.exists(miniscope_path):  # Check if Miniscope folder exists
                            sp_folder_path = os.path.join(dirpath, dirname, 'SP')
                            if not os.path.exists(sp_folder_path):
                                os.makedirs(sp_folder_path)
                                print(f"Created folder: {sp_folder_path}")

                            # Move content from Miniscope to SP folder
                            for item in os.listdir(miniscope_path):
                                source_item = os.path.join(miniscope_path, item)
                                dest_item = os.path.join(sp_folder_path, item)
                                shutil.move(source_item, dest_item)
                            print(f"Moved data from {miniscope_path} to {sp_folder_path}")
```

# rename the animal name
- List all the direct subdirectories of base_path.
- Rename each subdirectory by prefixing it with 'JH_' (If not added yet) and suffixing with '_n'.

```python
import os

base_path = r'H:\Miniscope_data\ACC_formalin_SP_data\naive'

# List all directories directly under base_path
subfolders = [d for d in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, d))]

for subfolder in subfolders:
    if not subfolder.startswith("JH_"):
        new_name = 'JH_' + subfolder + '_n'
        os.rename(os.path.join(base_path, subfolder), os.path.join(base_path, new_name))
        print(f"Renamed {subfolder} to {new_name}")
```
